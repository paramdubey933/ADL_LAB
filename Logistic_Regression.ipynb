```python
#removing setosa samples from the dataset, since logistic regr. gives binary classification
df=df[df['species']!='setosa']
```

```python
df.shape
```

```python
#convert the categorical features into numerical features 
df['species']=df['species'].map({'versicolor':0,'virginica':1})
```

```python
X=df.iloc[:,:-1] #IV-Independent Variables, all rows, columns except the last columns
y=df.iloc[:,-1] #DV-Dependent Variables, all rows and only the last column
```

```python
X.shape, y.shape # y.shape cant be (100,1) because it becomes a vector i.e 1D
```

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

```python
classifier.fit(X_train,y_train)
```

```python
y_pred=classifier.predict(X_test)
```

```python
from sklearn.metrics import(
confusion_matrix,
accuracy_score,
classification_report,
recall_score,
f1_score,
log_loss)
```

```python
acc=accuracy_score(y_test,y_pred)
acc
```

```python
f1=f1_score(y_test,y_pred)
f1
```

```python
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
print(classification_report(y_test,y_pred))
```
